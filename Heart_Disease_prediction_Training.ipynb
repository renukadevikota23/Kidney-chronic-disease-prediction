```python
import pandas as pd
dt=pd.read_csv('/content/heart (2).csv')
dt.head()
```

```python
dt=dt.drop(['chol','fbs'],axis=1)
```

```python
dt.head()
```

```python
x = dt.iloc[:,0:11]
y = dt['target']
```

```python
x
```

```python
y
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x.shape,y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

add - adds into neural network       
dense - create a neural network

```python
#create a nural network
model.add(Dense(units = 11,activation='relu'))
model.add(Dense(units = 5,activation='relu'))
model.add(Dense(units = 1,activation='sigmoid'))
```

relu stands for rectified linear unit
it takes only positive values and eliminate negitive values to get the precise result

sigmoid- it will give the perfect output without any modifications


```python
model.build(input_shape=(None,11))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
model.save('heart.h5')
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score  #Evaluation
accuracy_score(y_test,y_pred)
```
